```python
from audio_utils import Sample

import matplotlib.pyplot as plt
import numpy as np

file = './datasets/ravdess/calm/03-01-02-01-02-02-24.wav'
```

```python
sample = Sample(file)
print (sample.get_bit_rate())
print (sample.get_sample_rate())
print (sample.get_seconds())
print (len(sample.get_data()))

x = np.linspace(0, sample.get_seconds(), len(sample.get_data()), endpoint=False)
y = sample.get_data()
plt.plot(x, y)
plt.legend('data', loc='best')
plt.show()
```

```python
sample.resample(4, 24000)
sample.remove_silence(silence_thresh=-90)

print (sample.get_bit_rate())
print (sample.get_sample_rate())
print (sample.get_seconds())
print (len(sample.get_data()))

x = np.linspace(0, sample.get_seconds(), len(sample.get_data()), endpoint=False)
y = sample.get_data()
plt.plot(x, y)
plt.legend('data', loc='best')
plt.show()
```
